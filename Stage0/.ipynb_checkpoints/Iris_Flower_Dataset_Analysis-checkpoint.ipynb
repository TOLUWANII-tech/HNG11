```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Define te colum names for each off the five columns given raw in the dataset
column_names = ['sepal length (cm)','sepal width (cm)', 'petal length (cm)', 'petal width (cm)', 'class']
df = pd.read_csv('iris.data', delimiter=',', names = column_names)
df
```

```python
df.info()
```

```python
df.describe()
```

```python
sns.histplot(df['class'])
plt.show()
```

#### From the figure above, it is clear to see that there are equal numbers of each of the three clases of flowers in the dataset

```python

sns.scatterplot(x='sepal length (cm)', y='sepal width (cm)', data = df, hue = 'class')
plt.savefig('sepal_length_and_width.jpg')
```

##### From this plot it becomes clear to see that, flowers with long, narrow sepals are most likely to be iris-setosa, while the iris-virginica have the widest sepals, with the iris-versicolor more likely to have ths shortest sepal amongst the amples.

```python
sns.jointplot(x='petal length (cm)', y='petal width (cm)', data = df, hue = 'class')
plt.savefig('petal_length_and_width.jpg')
```

##### There exists a linear relationhip between petal length and width of the flowers. With iris-setora flowers having the shortest and narrowest petals while iris virginica flowers hae the longest and widet petals.

```python
sns.jointplot(x='petal length (cm)', y='sepal length (cm)', data = df, hue = 'class')
plt.savefig('petal_length_sepal_length.jpg')
```
